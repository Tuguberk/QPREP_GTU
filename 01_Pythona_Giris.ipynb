## 2.1 NumPy Dizileri (Arrays) Oluşturma

NumPy dizileri, birden fazla sayıyı verimli bir şekilde tutan yapılardır. Önce tanıdık örneklerle başlayalım:

## 2.2 Matris Çarpımı (Dot Product)

Matris çarpımı, birçok alanda karşımıza çıkan temel bir işlemdir. `@` operatörü veya `np.dot()` fonksiyonu ile yapılır.

Gündelik örneklerle anlayalım:

````python
# --- Gündelik Örnek: Alışveriş Hesabı ---
# Bir marketten aldığınız ürünlerin fiyatları (TL)
fiyatlar = np.array([3.50, 12.00, 8.75, 2.25])  # ekmek, peynir, zeytin, su

# Her üründen kaç tane aldınız
miktarlar = np.array([2, 1, 1, 3])  # 2 ekmek, 1 peynir, 1 zeytin, 3 su

# Toplam tutar = fiyatlar · miktarlar (dot product!)
toplam = np.dot(fiyatlar, miktarlar)
# veya: toplam = fiyatlar @ miktarlar

print("Ürünler:    ekmek  peynir  zeytin  su")
print("Fiyatlar:", fiyatlar)
print("Miktarlar:", miktarlar)
print(f"\nToplam tutar: {toplam:.2f} TL")
print(f"  → (3.50×2) + (12.00×1) + (8.75×1) + (2.25×3) = {toplam:.2f} TL")
print("\n💡 İşte dot product budur: karşılıklı elemanları çarp, sonuçları topla!")
````

````python
# --- Gündelik Örnek: Matris-Vektör Çarpımı ---
# 3 farklı market, 2 ürünün fiyatları
#                   ekmek  süt
market_fiyatlari = np.array([
    [3.50, 25.00],   # Market A
    [3.75, 22.50],   # Market B
    [3.25, 27.00]    # Market C
])

# Almak istediğimiz miktarlar
alisveris = np.array([2, 1])  # 2 ekmek, 1 süt

# Her marketteki toplam tutar (matris @ vektör)
tutarlar = market_fiyatlari @ alisveris

print("Her marketteki toplam tutar:")
marketler = ['Market A', 'Market B', 'Market C']
for market, tutar in zip(marketler, tutarlar):
    print(f"  {market}: {tutar:.2f} TL")
print(f"\n→ En uygun: {marketler[tutarlar.argmin()]} ({tutarlar.min():.2f} TL)")
# tutarlar.argmin() ile en ucuz fiyatın indeksini buluyor (Örn: 1).
````

````python
# Bazı kullanışlı NumPy fonksiyonları
print("Sıfır vektörü:", np.zeros(3))
print("Bir vektörü:", np.ones(4))
print("2×2 Birim matris:")
print(np.eye(2))
print("\nAralık (0 ile 1 arası, 5 eleman):", np.linspace(0, 1, 5))
````

## 4.1 İlk Kuantum Devremiz: 1 Kubit

Kuantum devreleri, kubitlere kapılar uygulayarak oluşturulur. Şimdi 1 kubitlik basit bir devre yapalım.

### Hadamard Kapısı (H-Gate) Nedir?

Hadamard kapısı, bir kubiti **süperpozisyon** durumuna getirir:

- Eğer kubit `|0⟩` durumundaysa → `|+⟩ = (|0⟩ + |1⟩) / √2` olur
- Bu, kubiti ölçtüğümüzde **%50 ihtimalle `0`**, **%50 ihtimalle `1`** sonucu alacağımız anlamına gelir
- Tıpkı havaya atılan adil bir yazı-tura gibi — ama kuantum mekaniğinin kurallarıyla! 🪙

````python
# 1 kubit ve 1 klasik bit içeren bir kuantum devresi oluştur
qc = QuantumCircuit(1, 1)

# Hadamard kapısını kubit 0'a uygula → süperpozisyon!
qc.h(0)

# Kubiti ölç ve sonucu klasik bite yaz
qc.measure(0, 0)

# Devreyi çizdir
print("İlk Kuantum Devreniz:")
print("=" * 40)
qc.draw(output='text')
````

````python
# Devreyi görsel olarak çizdirelim (Matplotlib ile)
qc_gorsel = QuantumCircuit(1, 1)
qc_gorsel.h(0)
qc_gorsel.measure(0, 0)

qc_gorsel.draw(output='mpl', style='iqp')
#qc_gorsel.draw(output='mpl', filename='devrem.png')
````

## 4.2 Devreyi Adım Adım Anlayalım

```
     ┌───┐┌─┐
  q: ┤ H ├┤M├
     └───┘└╥┘
  c:  ═════╩═
```

| Adım | İşlem | Açıklama |
|:---:|:---|:---|
| 1 | `QuantumCircuit(1, 1)` | 1 kubit (`q`) ve 1 klasik bit (`c`) oluştur |
| 2 | `qc.h(0)` | Kubit 0'a Hadamard kapısı uygula → süperpozisyon |
| 3 | `qc.measure(0, 0)` | Kubit 0'ı ölç, sonucu klasik bit 0'a yaz |

> ⚛️ **Bu devre ne yapıyor?** Kubiti `|0⟩` başlangıç durumundan alıp `|+⟩` süperpozisyon durumuna getiriyor. Ölçüm yaptığımızda %50 olasılıkla `0`, %50 olasılıkla `1` sonucu alıyoruz. Bu, kuantum mekaniğinin olasılıksal doğasının en basit gösterimidir!

## 4.3 Devreyi Simüle Edelim! 🚀

Devremizi oluşturduk — şimdi sıra **çalıştırmaya** geldi! Gerçek bir kuantum bilgisayar yerine, bilgisayarımızdaki **simülatör** kullanarak devreyi çalıştıracağız.

### Simülasyon Nedir?
- Kuantum devresi klasik bilgisayarda **taklit edilir** (simüle edilir)
- Qiskit'in `AerSimulator` bileşeni bu işi yapar
- `shots` parametresi: Devreyi **kaç kez** tekrarlayacağımızı belirler
- Her çalıştırmada kubit farklı bir sonuç verebilir (kuantum olasılıksal!)

> 🪙 Bunu yazı-tura atmaya benzetebilirsiniz: 1 kez atarsanız `0` veya `1` görürsünüz. Ama 1024 kez atarsanız, yaklaşık **%50 yazı, %50 tura** çıktığını görürsünüz!

````python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# --- ADIM 1: Devreyi oluştur ---
qc = QuantumCircuit(1, 1)
qc.h(0)           # Hadamard kapısı → süperpozisyon
qc.measure(0, 0)  # Ölçüm

# --- ADIM 2: Simülatörü seç ---
simulator = AerSimulator()

# --- ADIM 3: Devreyi simülatör için derle (transpile) ---
transpiled_qc = transpile(qc, simulator)

# --- ADIM 4: Simülasyonu çalıştır (1024 tekrar) ---
result = simulator.run(transpiled_qc, shots=1024).result()

# --- ADIM 5: Sonuçları al ---
counts = result.get_counts()

print("🎲 Simülasyon Sonuçları (1024 ölçüm):")
print("=" * 40)
print(f"  Ölçüm sonuçları: {counts}")
print()
for durum, sayi in sorted(counts.items()):
    yuzde = sayi / 1024 * 100
    bar = '█' * int(yuzde / 2)
    print(f"  |{durum}⟩ : {sayi:4d} kez  ({yuzde:5.1f}%)  {bar}")
print()
print("→ Sonuçlar yaklaşık %50 / %50 — süperpozisyon doğrulandı! ✅")
````

## 4.4 Sonuçları Görselleştirelim 📊

Simülasyon sonuçlarını hem Qiskit'in kendi histogram aracıyla hem de Matplotlib ile görselleştirelim.

````python
# --- Qiskit'in Kendi Histogram Aracı ---
from qiskit.visualization import plot_histogram

plot_histogram(
    counts, 
    title='Hadamard Devresi — 1024 Ölçüm Sonucu',
    color='#6C5CE7',
    figsize=(8, 5)
)
````

````python
import matplotlib.pyplot as plt
import numpy as np

# --- Matplotlib ile Özel Görselleştirme ---
durumlar = sorted(counts.keys())
sayimlar = [counts[d] for d in durumlar]
yüzdeler = [s / sum(sayimlar) * 100 for s in sayimlar]
etiketler = [f'|{d}⟩' for d in durumlar]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# --- Sol: Bar Grafiği ---
renkler = ['#00B894', '#E17055']
bars = ax1.bar(etiketler, sayimlar, color=renkler, edgecolor='white', 
               linewidth=2, width=0.5, zorder=3)

# Çubukların üzerine sayı ve yüzde yaz
for i, (bar, sayim, yzd) in enumerate(zip(bars, sayimlar, yüzdeler)):
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 15,
             f'{sayim}\n(%{yzd:.1f})', ha='center', fontsize=13, fontweight='bold')

# %50 referans çizgisi
ax1.axhline(y=512, color='#636e72', linestyle='--', linewidth=1.5, 
            alpha=0.7, label='Beklenen: 512 (tam %50)')

ax1.set_xlabel('Kubit Durumu', fontsize=13)
ax1.set_ylabel('Ölçüm Sayısı', fontsize=13)
ax1.set_title('Ölçüm Sonuçları (Sayı)', fontsize=14, fontweight='bold')
ax1.set_ylim(0, max(sayimlar) + 80)
ax1.legend(fontsize=10)
ax1.grid(axis='y', alpha=0.3)

# --- Sağ: Pasta Grafiği ---
pasta_renkler = ['#00B894', '#E17055']
wedges, texts, autotexts = ax2.pie(
    sayimlar, labels=etiketler, colors=pasta_renkler,
    autopct='%1.1f%%', startangle=90,
    textprops={'fontsize': 14, 'fontweight': 'bold'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
    explode=(0.03, 0.03)
)
for autotext in autotexts:
    autotext.set_fontsize(15)
    autotext.set_fontweight('bold')

ax2.set_title('Olasılık Dağılımı (%)', fontsize=14, fontweight='bold')

fig.suptitle('⚛️ Hadamard Kapısı — Süperpozisyon Simülasyonu (1024 shot)', 
             fontsize=16, fontweight='bold', y=1.02)

plt.tight_layout()
plt.show()

print("\n🎯 Sonuç: Hadamard kapısı kubiti tam süperpozisyona getirdi!")
print("   Her ölçümde |0⟩ veya |1⟩ alma olasılığımız yaklaşık eşit (~%50 / ~%50).")
print("   Bu, kuantum mekaniğinin temel ilkelerinden birinin canlı gösterimidir!")
````

> 🔍 **Neden tam olarak %50 değil?** Kuantum mekaniği *olasılıksal* bir teoridir. Her ölçüm rastgeledir. Tıpkı para atışı gibi — 1024 kez atarsanız tam 512-512 çıkması gerekmez, ama sonuçlar %50'ye yakın olacaktır. Deneyi birkaç kez çalıştırarak farklı sonuçlar elde edebilirsiniz!

> 💡 **Deneyin:** Yukarıdaki hücredeki `shots=1024` değerini `shots=100` veya `shots=10000` yaparak tekrar çalıştırın. Daha fazla ölçüm = %50'ye daha yakın sonuçlar!

## 4.5 NumPy ile Doğrulama

Az önce NumPy ile öğrendiğimiz matris çarpımını kullanarak bu devrenin ne yaptığını **matematiksel olarak** doğrulayalım:

````python
import numpy as np

# |0⟩ başlangıç durumu
ket_0 = np.array([1, 0])

# Hadamard matrisi
H = np.array([[1, 1], [1, -1]]) / np.sqrt(2)

# H kapısını uygula
sonuc = H @ ket_0

print("Başlangıç durumu |0⟩ :", ket_0)
print("H kapısı sonrası |+⟩ :", sonuc)
print()
print("Ölçüm olasılıkları:")
print(f"  P(|0⟩) = |{sonuc[0]:.4f}|² = {abs(sonuc[0])**2:.2f}  →  %{abs(sonuc[0])**2 * 100:.0f}")
print(f"  P(|1⟩) = |{sonuc[1]:.4f}|² = {abs(sonuc[1])**2:.2f}  →  %{abs(sonuc[1])**2 * 100:.0f}")
print()
print("✅ Tam olarak beklediğimiz gibi — %50 / %50 eşit olasılık!")
print()
print("📐 Teori (NumPy) ve pratiğin (Qiskit simülasyonu) uyumunu gördük!")
print("   Matris çarpımı bize kesin olasılıkları verir (%50/%50),")
print("   simülasyon ise buna yakın ama rastgele sonuçlar üretir.")
````
