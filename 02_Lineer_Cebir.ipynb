# Bölüm 1: Vektörler

## Vektör Nedir?

Vektör, sıralı sayılar dizisidir. Kuantum hesaplamada:
- **Kubit durumları** vektörlerle temsil edilir
- `|0⟩ = [1, 0]` ve `|1⟩ = [0, 1]` en temel kuantum durumlarıdır

Matematiksel gösterim:

$$\vec{v} = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix}$$

### Temel Vektör İşlemleri
1. **Toplama**: Karşılıklı elemanları topla
2. **Skaler Çarpım**: Her elemanı bir sayı ile çarp
3. **Büyüklük (Norm)**: Vektörün uzunluğu — $||\vec{v}|| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}$

## 1.1 Vektör Tanımlama ve Toplama

### 🐍 Saf Python ile Vektörler

```python
# === SAF PYTHON İLE VEKTÖRLER ===

# Vektörleri Python listesi olarak tanımlayalım
v1 = [2, 3, 5]
v2 = [1, 4, -2]

print("v1 =", v1)
print("v2 =", v2)
print()

# --- Vektör Toplama ---
# Karşılıklı elemanları teker teker topluyoruz
def vektor_topla(a, b):
    """İki vektörü toplar (eleman bazlı)."""
    sonuc = []
    for i in range(len(a)):
        sonuc.append(a[i] + b[i])
    return sonuc

toplam = vektor_topla(v1, v2)
print("v1 + v2 =", toplam)
print(f"  → [{v1[0]}+{v2[0]}, {v1[1]}+{v2[1]}, {v1[2]}+{v2[2]}] = {toplam}")
print()

# --- Skaler Çarpım ---
# Her elemanı bir sayı (skaler) ile çarpıyoruz
def skaler_carp(k, v):
    """Bir vektörü skaler k ile çarpar."""
    return [k * vi for vi in v]

k = 3
sonuc_skaler = skaler_carp(k, v1)
print(f"{k} × v1 = {sonuc_skaler}")
print(f"  → [{k}×{v1[0]}, {k}×{v1[1]}, {k}×{v1[2]}] = {sonuc_skaler}")
print()

# --- Vektör Büyüklüğü (Norm) ---
import math

def vektor_norm(v):
    """Vektörün büyüklüğünü (L2 norm) hesaplar."""
    toplam_kare = 0
    for vi in v:
        toplam_kare += vi ** 2
    return math.sqrt(toplam_kare)

norm_v1 = vektor_norm(v1)
print(f"||v1|| = √({v1[0]}² + {v1[1]}² + {v1[2]}²) = √{sum(x**2 for x in v1)} ≈ {norm_v1:.4f}")
```

### 📊 NumPy ile Vektörler

```python
# === NUMPY İLE VEKTÖRLER ===

# NumPy array olarak tanımlama
v1 = np.array([2, 3, 5])
v2 = np.array([1, 4, -2])

print("v1 =", v1)
print("v2 =", v2)
print()

# --- Vektör Toplama --- (tek satırda!)
toplam = v1 + v2
print("v1 + v2 =", toplam)
print()

# --- Skaler Çarpım --- (tek satırda!)
k = 3
print(f"{k} × v1 =", k * v1)
print()

# --- Vektör Büyüklüğü (Norm) ---
norm_v1 = np.linalg.norm(v1)
print(f"||v1|| = {norm_v1:.4f}")
print()

# --- Kuantum Bağlantısı: Kubit durumları ---
ket_0 = np.array([1, 0])  # |0⟩ durumu
ket_1 = np.array([0, 1])  # |1⟩ durumu

print("⚛️ Kuantum durumları da birer vektördür:")
print(f"  |0⟩ = {ket_0}")
print(f"  |1⟩ = {ket_1}")
print(f"  ||0⟩| = {np.linalg.norm(ket_0):.1f}  (birim vektör ✅)")
print(f"  ||1⟩| = {np.linalg.norm(ket_1):.1f}  (birim vektör ✅)")

print("\n💡 NumPy ile vektör işlemleri çok daha kısa ve okunabilir!")
```

# Bölüm 2: Matrisler

## Matris Nedir?

Matris, sayıların satır ve sütunlar halinde düzenlenmiş dikdörtgensel bir tablosudur.

$$A = \begin{bmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{bmatrix}$$

Kuantum hesaplamada:
- **Kuantum kapıları** matrislerle temsil edilir
- Hadamard kapısı: $H = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}$
- Pauli-X kapısı: $X = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$

### Temel Matris İşlemleri
1. **Toplama**: Karşılıklı elemanları topla
2. **Skaler Çarpım**: Her elemanı bir sayı ile çarp
3. **Transpoz**: Satır ve sütunları yer değiştir — $A^T$

## 2.1 Matris Tanımlama ve İşlemler

### 🐍 Saf Python ile Matrisler

```python
# === SAF PYTHON İLE MATRİSLER ===

# Matrisleri iç içe listeler (liste içinde liste) olarak tanımlıyoruz
# Her iç liste bir satırı temsil eder
A = [
    [1, 2, 3],
    [4, 5, 6]
]

B = [
    [7, 8, 9],
    [10, 11, 12]
]

def matris_yazdir(ad, M):
    """Matrisi güzel formatta yazdırır."""
    print(f"{ad} =")
    for satir in M:
        print("  ", satir)
    print()

matris_yazdir("A", A)
matris_yazdir("B", B)

# --- Matris Toplama ---
def matris_topla(A, B):
    """İki matrisi toplar (eleman bazlı)."""
    satir_sayisi = len(A)
    sutun_sayisi = len(A[0])
    sonuc = []
    for i in range(satir_sayisi):
        satir = []
        for j in range(sutun_sayisi):
            satir.append(A[i][j] + B[i][j])
        sonuc.append(satir)
    return sonuc

C = matris_topla(A, B)
matris_yazdir("A + B", C)

# --- Matris Transpoz ---
def matris_transpoz(M):
    """Matrisin transpozunu alır (satır ↔ sütun)."""
    satir_sayisi = len(M)
    sutun_sayisi = len(M[0])
    sonuc = []
    for j in range(sutun_sayisi):
        yeni_satir = []
        for i in range(satir_sayisi):
            yeni_satir.append(M[i][j])
        sonuc.append(yeni_satir)
    return sonuc

A_T = matris_transpoz(A)
print("A matrisi (2×3):")
matris_yazdir("A", A)
print("A'nın transpozu (3×2):")
matris_yazdir("Aᵀ", A_T)

# --- Skaler Çarpım ---
def matris_skaler_carp(k, M):
    """Matrisi skaler k ile çarpar."""
    sonuc = []
    for satir in M:
        yeni_satir = [k * eleman for eleman in satir]
        sonuc.append(yeni_satir)
    return sonuc

k = 2
matris_yazdir(f"{k} × A", matris_skaler_carp(k, A))
```

### 📊 NumPy ile Matrisler

```python
# === NUMPY İLE MATRİSLER ===

A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8, 9],
    [10, 11, 12]
])

print("A =")
print(A)
print(f"Boyut: {A.shape} (satır × sütun)")
print()

# --- Matris Toplama ---
print("A + B =")
print(A + B)
print()

# --- Skaler Çarpım ---
print("2 × A =")
print(2 * A)
print()

# --- Transpoz ---
print(f"A boyutu: {A.shape}")
print("A =")
print(A)
print()
print(f"Aᵀ boyutu: {A.T.shape}")
print("Aᵀ =")
print(A.T)
print()

# --- Kuantum Bağlantısı: Kuantum Kapıları ---
print("=" * 40)
print("⚛️ Kuantum kapıları da matrislerdir:")
print()

# Pauli-X kapısı (NOT kapısı)
X = np.array([[0, 1], [1, 0]])
print("Pauli-X (NOT kapısı):")
print(X)
print()

# Hadamard kapısı
H = np.array([[1, 1], [1, -1]]) / np.sqrt(2)
print("Hadamard kapısı (H):")
print(H.round(4))

print("\n💡 NumPy ile matris işlemleri tek satırda yapılabilir!")
```

# Bölüm 3: Dot Product (İç Çarpım)

## Dot Product Nedir?

İki vektörün **karşılıklı elemanlarını çarpıp toplama** işlemidir. Sonuç tek bir **sayıdır** (skaler).

$$\vec{a} \cdot \vec{b} = a_1 b_1 + a_2 b_2 + \cdots + a_n b_n = \sum_{i=1}^{n} a_i b_i$$

### Örnek:
$$\begin{bmatrix} 2 \\ 3 \end{bmatrix} \cdot \begin{bmatrix} 4 \\ 5 \end{bmatrix} = (2 \times 4) + (3 \times 5) = 8 + 15 = 23$$

### Kuantum Bağlantısı:
- Kuantum mekaniğinde ölçüm olasılıkları **iç çarpım** ile hesaplanır
- İki kuantum durumunun **ortogonalliği** (birbirine dik olması) iç çarpım ile belirlenir
- $\langle 0 | 1 \rangle = 0$ → |0⟩ ve |1⟩ birbirine diktir (ortogonal)

### 🐍 Saf Python ile Dot Product

```python
# === SAF PYTHON İLE DOT PRODUCT ===

def dot_product(a, b):
    """İki vektörün iç çarpımını hesaplar."""
    if len(a) != len(b):
        raise ValueError("Vektörler aynı boyutta olmalıdır!")
    
    toplam = 0
    for i in range(len(a)):
        toplam += a[i] * b[i]
    return toplam

# Örnek 1: Basit vektörler
a = [2, 3]
b = [4, 5]

sonuc = dot_product(a, b)
print(f"a = {a}")
print(f"b = {b}")
print(f"a · b = ({a[0]}×{b[0]}) + ({a[1]}×{b[1]}) = {a[0]*b[0]} + {a[1]*b[1]} = {sonuc}")
print()

# Örnek 2: 3 boyutlu vektörler
c = [1, -2, 3]
d = [4, 0, -1]

sonuc2 = dot_product(c, d)
print(f"c = {c}")
print(f"d = {d}")
print(f"c · d = ({c[0]}×{d[0]}) + ({c[1]}×{d[1]}) + ({c[2]}×{d[2]})")
print(f"      = {c[0]*d[0]} + {c[1]*d[1]} + {c[2]*d[2]} = {sonuc2}")
print()

# Örnek 3: Ortogonal (dik) vektörler
e = [1, 0]
f = [0, 1]

sonuc3 = dot_product(e, f)
print(f"e = {e}  (|0⟩ kubit durumu)")
print(f"f = {f}  (|1⟩ kubit durumu)")
print(f"e · f = {sonuc3}  → Ortogonal (dik)! ✅")
print("💡 İç çarpım 0 ise vektörler birbirine diktir.")
```

### 📊 NumPy ile Dot Product

```python
# === NUMPY İLE DOT PRODUCT ===

a = np.array([2, 3])
b = np.array([4, 5])

# Yöntem 1: np.dot()
sonuc1 = np.dot(a, b)
print(f"np.dot(a, b) = {sonuc1}")

# Yöntem 2: @ operatörü
sonuc2 = a @ b
print(f"a @ b        = {sonuc2}")
print()

# --- Kuantum Bağlantısı: Ortogonallik ---
ket_0 = np.array([1, 0])  # |0⟩
ket_1 = np.array([0, 1])  # |1⟩

print("⚛️ Kubit durumlarının ortogonalliği:")
print(f"  ⟨0|0⟩ = {np.dot(ket_0, ket_0)}  (kendisiyle → 1)")
print(f"  ⟨1|1⟩ = {np.dot(ket_1, ket_1)}  (kendisiyle → 1)")
print(f"  ⟨0|1⟩ = {np.dot(ket_0, ket_1)}  (birbirine dik → 0)")
print(f"  ⟨1|0⟩ = {np.dot(ket_1, ket_0)}  (birbirine dik → 0)")
print()

# Süperpozisyon durumu
ket_plus = np.array([1, 1]) / np.sqrt(2)  # |+⟩ = (|0⟩ + |1⟩) / √2
print(f"  |+⟩ = {ket_plus.round(4)}")
print(f"  ⟨+|0⟩ = {np.dot(ket_plus, ket_0):.4f}  → |0⟩ ölçme olasılığı: {np.dot(ket_plus, ket_0)**2:.2f} (%50)")
print(f"  ⟨+|1⟩ = {np.dot(ket_plus, ket_1):.4f}  → |1⟩ ölçme olasılığı: {np.dot(ket_plus, ket_1)**2:.2f} (%50)")
```

# Bölüm 4: Matris Çarpımı

## Matris Çarpımı Nedir?

İki matrisin çarpımında, **birinci matrisin satırları** ile **ikinci matrisin sütunları** arasında dot product yapılır.

### Boyut Kuralı
- $A$ matrisi: $m \times n$ boyutunda
- $B$ matrisi: $n \times p$ boyutunda
- Sonuç $C = A \times B$: $m \times p$ boyutunda

⚠️ **A'nın sütun sayısı = B'nin satır sayısı** olmalıdır! Aksi halde çarpım tanımsızdır.

### Formül:
$$C_{ij} = \sum_{k=1}^{n} A_{ik} \cdot B_{kj}$$

### Kuantum Bağlantısı:
Bir kuantum kapısını (matris) bir kubit durumuna (vektör) uygulamak = **matris-vektör çarpımı**

$$|\psi'\rangle = U \cdot |\psi\rangle$$

### 🐍 Saf Python ile Matris Çarpımı

```python
# === SAF PYTHON İLE MATRİS ÇARPIMI ===

def matris_carp(A, B):
    """İki matrisi çarpar. A: m×n, B: n×p → Sonuç: m×p"""
    m = len(A)       # A'nın satır sayısı
    n = len(A[0])    # A'nın sütun sayısı
    p = len(B[0])    # B'nin sütun sayısı
    
    # Boyut kontrolü
    if n != len(B):
        raise ValueError(f"Boyutlar uyumsuz: A ({m}×{n}) ve B ({len(B)}×{p})")
    
    # Sonuç matrisi (m×p boyutunda sıfırlarla başlat)
    C = [[0 for _ in range(p)] for _ in range(m)]
    
    for i in range(m):       # A'nın her satırı
        for j in range(p):   # B'nin her sütunu
            for k in range(n):  # Karşılıklı çarp ve topla
                C[i][j] += A[i][k] * B[k][j]
    
    return C

# --- Matris × Matris ---
A = [
    [1, 2],
    [3, 4]
]

B = [
    [5, 6],
    [7, 8]
]

C = matris_carp(A, B)

print("A =")
for satir in A:
    print(" ", satir)
print()

print("B =")
for satir in B:
    print(" ", satir)
print()

print("A × B =")
for satir in C:
    print(" ", satir)
print()

# Adım adım açıklama
print("Adım adım hesaplama:")
print(f"  C[0][0] = ({A[0][0]}×{B[0][0]}) + ({A[0][1]}×{B[1][0]}) = {A[0][0]*B[0][0]} + {A[0][1]*B[1][0]} = {C[0][0]}")
print(f"  C[0][1] = ({A[0][0]}×{B[0][1]}) + ({A[0][1]}×{B[1][1]}) = {A[0][0]*B[0][1]} + {A[0][1]*B[1][1]} = {C[0][1]}")
print(f"  C[1][0] = ({A[1][0]}×{B[0][0]}) + ({A[1][1]}×{B[1][0]}) = {A[1][0]*B[0][0]} + {A[1][1]*B[1][0]} = {C[1][0]}")
print(f"  C[1][1] = ({A[1][0]}×{B[0][1]}) + ({A[1][1]}×{B[1][1]}) = {A[1][0]*B[0][1]} + {A[1][1]*B[1][1]} = {C[1][1]}")
print()

# --- Matris × Vektör ---
def matris_vektor_carp(M, v):
    """Matrisi vektör ile çarpar. M: m×n, v: n boyutlu → Sonuç: m boyutlu"""
    m = len(M)
    n = len(M[0])
    sonuc = [0] * m
    for i in range(m):
        for j in range(n):
            sonuc[i] += M[i][j] * v[j]
    return sonuc

# Pauli-X kapısını |0⟩'a uygulayalım
X = [[0, 1], [1, 0]]  # NOT kapısı
ket_0 = [1, 0]

sonuc = matris_vektor_carp(X, ket_0)
print("⚛️ Kuantum: X kapısını |0⟩'a uygulama:")
print(f"  X × |0⟩ = {sonuc}  → |1⟩ durumu! (bit flip ✅)")
```

### 📊 NumPy ile Matris Çarpımı

```python
# === NUMPY İLE MATRİS ÇARPIMI ===

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Yöntem 1: @ operatörü (önerilen)
C = A @ B
print("A @ B =")
print(C)
print()

# Yöntem 2: np.matmul()
C2 = np.matmul(A, B)
print("np.matmul(A, B) =")
print(C2)
print()

# --- Kuantum Kapılarını Uygulama ---
print("=" * 40)
print("⚛️ Kuantum kapılarının uygulanması:")
print()

# Pauli-X: |0⟩ → |1⟩
X = np.array([[0, 1], [1, 0]])
ket_0 = np.array([1, 0])
sonuc_x = X @ ket_0
print(f"X × |0⟩ = {sonuc_x}  → |1⟩ (NOT kapısı ✅)")
print()

# Hadamard: |0⟩ → |+⟩
H = np.array([[1, 1], [1, -1]]) / np.sqrt(2)
sonuc_h = H @ ket_0
print(f"H × |0⟩ = {sonuc_h.round(4)}  → |+⟩ (süperpozisyon ✅)")
print()

# İki kapıyı art arda uygulama: H sonra X
sonuc_hx = X @ (H @ ket_0)  # Önce H, sonra X
print(f"X × H × |0⟩ = {sonuc_hx.round(4)}")
print()

# Birim matris ile çarpım (değişmez)
I = np.eye(2)
print(f"I × |0⟩ = {(I @ ket_0)}  → Değişmedi (birim matris ✅)")

print("\n💡 '@' operatörü NumPy'da matris çarpımı için en pratik yoldur!")
```

# Bölüm 5: Tensor Product (Tensör Çarpımı / Kronecker Çarpımı)

## Tensor Product Nedir?

Tensor çarpımı (Kronecker çarpımı), iki matrisi veya vektörü birleştirerek **daha büyük boyutlu** bir yapı oluşturur.

### Vektörlerde Tensor Product:

$$\begin{bmatrix} a_1 \\ a_2 \end{bmatrix} \otimes \begin{bmatrix} b_1 \\ b_2 \end{bmatrix} = \begin{bmatrix} a_1 \cdot b_1 \\ a_1 \cdot b_2 \\ a_2 \cdot b_1 \\ a_2 \cdot b_2 \end{bmatrix}$$

### Matrislerde Tensor Product:

$$A \otimes B = \begin{bmatrix} a_{11} B & a_{12} B \\ a_{21} B & a_{22} B \end{bmatrix}$$

Her eleman, skaler ile **B matrisinin tamamının** çarpımı olur.

### Kuantum Bağlantısı:
- **Çoklu kubit sistemleri** tensor çarpımı ile ifade edilir
- 2 kubitlik bir sistem: $|\psi\rangle = |q_1\rangle \otimes |q_2\rangle$
- Örnek: $|00\rangle = |0\rangle \otimes |0\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix} \otimes \begin{bmatrix} 1 \\ 0 \end{bmatrix} = \begin{bmatrix} 1 \\ 0 \\ 0 \\ 0 \end{bmatrix}$

### Boyut Kuralı:
- $A$: $m \times n$, $B$: $p \times q$ → $A \otimes B$: $(m \cdot p) \times (n \cdot q)$
- Vektörler için: $a$: $m$ boyutlu, $b$: $n$ boyutlu → $a \otimes b$: $m \cdot n$ boyutlu

### 🐍 Saf Python ile Tensor Product

```python
# === SAF PYTHON İLE TENSOR PRODUCT ===

# --- Vektörlerde Tensor Product ---
def vektor_tensor_product(a, b):
    """İki vektörün tensor çarpımını hesaplar."""
    sonuc = []
    for ai in a:
        for bi in b:
            sonuc.append(ai * bi)
    return sonuc

# Kubit durumlarının tensor çarpımı
ket_0 = [1, 0]  # |0⟩
ket_1 = [0, 1]  # |1⟩

# |00⟩ = |0⟩ ⊗ |0⟩
ket_00 = vektor_tensor_product(ket_0, ket_0)
print(f"|0⟩ = {ket_0}")
print(f"|0⟩ ⊗ |0⟩ = |00⟩ = {ket_00}")
print()

# |01⟩ = |0⟩ ⊗ |1⟩
ket_01 = vektor_tensor_product(ket_0, ket_1)
print(f"|0⟩ ⊗ |1⟩ = |01⟩ = {ket_01}")
print()

# |10⟩ = |1⟩ ⊗ |0⟩
ket_10 = vektor_tensor_product(ket_1, ket_0)
print(f"|1⟩ ⊗ |0⟩ = |10⟩ = {ket_10}")
print()

# |11⟩ = |1⟩ ⊗ |1⟩
ket_11 = vektor_tensor_product(ket_1, ket_1)
print(f"|1⟩ ⊗ |1⟩ = |11⟩ = {ket_11}")
print()

print("=" * 50)
print("📊 2 kubitlik tüm temel durumlar:")
print(f"  |00⟩ = {ket_00}")
print(f"  |01⟩ = {ket_01}")
print(f"  |10⟩ = {ket_10}")
print(f"  |11⟩ = {ket_11}")
print()

# --- Matrislerde Tensor Product ---
print("=" * 50)
print("\n--- Matrislerde Tensor Product ---")

def matris_tensor_product(A, B):
    """İki matrisin Kronecker (tensor) çarpımını hesaplar."""
    m_A, n_A = len(A), len(A[0])
    m_B, n_B = len(B), len(B[0])
    
    # Sonuç boyutu: (m_A * m_B) × (n_A * n_B)
    sonuc = [[0 for _ in range(n_A * n_B)] for _ in range(m_A * m_B)]
    
    for i in range(m_A):
        for j in range(n_A):
            for k in range(m_B):
                for l in range(n_B):
                    sonuc[i * m_B + k][j * n_B + l] = A[i][j] * B[k][l]
    
    return sonuc

# I ⊗ X (Birim matris ⊗ Pauli-X)
I = [[1, 0], [0, 1]]
X = [[0, 1], [1, 0]]

IX = matris_tensor_product(I, X)
print("\nI ⊗ X =")
for satir in IX:
    print(" ", satir)
```

### 📊 NumPy ile Tensor Product

```python
# === NUMPY İLE TENSOR PRODUCT ===

# np.kron() fonksiyonu Kronecker (tensor) çarpımını hesaplar

ket_0 = np.array([1, 0])  # |0⟩
ket_1 = np.array([0, 1])  # |1⟩

# --- Vektörlerde ---
print("⚛️ 2 kubitlik temel durumlar (tensor product):")
print(f"  |00⟩ = |0⟩ ⊗ |0⟩ = {np.kron(ket_0, ket_0)}")
print(f"  |01⟩ = |0⟩ ⊗ |1⟩ = {np.kron(ket_0, ket_1)}")
print(f"  |10⟩ = |1⟩ ⊗ |0⟩ = {np.kron(ket_1, ket_0)}")
print(f"  |11⟩ = |1⟩ ⊗ |1⟩ = {np.kron(ket_1, ket_1)}")
print()

# --- Matrislerde ---
print("⚛️ 2 kubitlik kuantum kapıları (tensor product):")
print()

I = np.eye(2)
X = np.array([[0, 1], [1, 0]])
H = np.array([[1, 1], [1, -1]]) / np.sqrt(2)

# I ⊗ X: İkinci kubite X uygula, birinciyi değiştirme
IX = np.kron(I, X)
print("I ⊗ X (sadece 2. kubite NOT uygula):")
print(IX.astype(int))
print()

# H ⊗ I: Birinci kubite H uygula, ikinciyi değiştirme
HI = np.kron(H, I)
print("H ⊗ I (sadece 1. kubite Hadamard uygula):")
print(HI.round(4))
print()

# H ⊗ H: Her iki kubite de Hadamard uygula
HH = np.kron(H, H)
print("H ⊗ H (her 2 kubite Hadamard uygula):")
print(HH.round(4))
print()

# --- Uygulama: |00⟩ durumuna H⊗H uygulama ---
ket_00 = np.kron(ket_0, ket_0)
sonuc = HH @ ket_00
print("=" * 50)
print("Uygulama: (H⊗H) × |00⟩")
print(f"  Başlangıç: |00⟩ = {ket_00}")
print(f"  Sonuç:            {sonuc.round(4)}")
print()
print("  Ölçüm olasılıkları:")
durumlar = ['|00⟩', '|01⟩', '|10⟩', '|11⟩']
for i, durum in enumerate(durumlar):
    olasilik = abs(sonuc[i]) ** 2
    print(f"    P({durum}) = |{sonuc[i]:.4f}|² = {olasilik:.2f}  →  %{olasilik*100:.0f}")
print()
print("  → Her 2 kubite H uyguladık: 4 durumun her biri eşit olasılıklı (%25)! ✅")

print("\n💡 np.kron() fonksiyonu tensor çarpımı için tek satırlık çözüm sunar!")
```

# 📝 Konu Özet Tablosu

| Konu | Saf Python | NumPy | Boyut Değişimi |
|:---|:---|:---|:---|
| **Vektör Toplama** | `for` döngüsü ile eleman bazlı | `v1 + v2` | (n,) + (n,) → (n,) |
| **Skaler Çarpım** | `[k * vi for vi in v]` | `k * v` | (n,) → (n,) |
| **Matris Toplama** | İç içe `for` döngüsü | `A + B` | (m,n) + (m,n) → (m,n) |
| **Transpoz** | Satır ↔ Sütun döngüsü | `A.T` | (m,n) → (n,m) |
| **Dot Product** | `sum(a[i]*b[i])` | `np.dot(a,b)` veya `a @ b` | (n,) · (n,) → skaler |
| **Matris Çarpımı** | 3 katlı `for` döngüsü | `A @ B` | (m,n) × (n,p) → (m,p) |
| **Tensor Product** | 2 katlı `for` (vektör), 4 katlı (matris) | `np.kron(A, B)` | (m,n) ⊗ (p,q) → (mp,nq) |

> 💡 **Önemli:** Saf Python örnekleri **mantığı anlamak** içindir. Gerçek projelerde her zaman **NumPy** kullanın — çok daha hızlı ve okunabilirdir!
